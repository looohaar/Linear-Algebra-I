**Solving Linear Systems**

1.Substitution Method

In the Substitution example, the two equations in the system are:

y = 3x

-5x + 2y = 2

substituting y value in 2nd equation

-5x + 2*3x =2, ie, 1x =2, x =2

Then y = 3x will be y = 3*2 ie, 6


```python
x= np.linspace(-10,10,1000) # start, finish, n points
```

```python
y1= 3*x
```

```python
y2=  1+ (5*x)/2
```

```python
fig, ax= plt.subplots()
plt.xlabel=('x')
plt.ylabel('y')
ax.set_xlim([0,3])
ax.set_ylim([0,10])
ax.plot(x, y1, c= 'green')
ax.plot(x ,y2, c='brown')
plt.axvline(x=2, color='purple', linestyle='--')

_ = plt.axhline(y=6, color='purple', linestyle='--')
```

2.Elimination method

In the Elimination example, the two equations in the system are:

2x -3y =15

4x + 10y =14

In elimination method we will eliminate any one of the variable and find  value of other. In this case we will eliminte the x value first, for tha multiply the first equation with 2 and subtract both the equations.

On solving the equation we will get (6,-1) as x and y values.


```python
y1= -5 + (2*x)/3
y2 = (7-2*x)/5
```

```python
fig, ax= plt.subplots()
plt.xlabel=('x')
plt.ylabel=('y')

# Add x and y axes:
plt.axvline(x=0, color= 'lightgray')
plt.axhline(y=0, color= 'lightgray')

ax.set_xlim([-2,10])
ax.set_ylim([-6,4])
ax.plot(x, y1, c= 'green')
ax.plot(x ,y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')

_ = plt.axhline(y=-1, color='purple', linestyle='--')
```

Matrix inverse of ***X*** is donoted as ***X***** -1, (***X*** rays to -1)

If we multiply matrix X with the inverse of X the result will be an Indetity matrix.



**IMP Note**

A **regression formula** can be derived from metrices, in this cas consider the example of a house price prediction model.

y = ***X*** w, (w is the vector of weight  a through m)

* y will the outcomes, ie, house prices

* ***X*** will be the features, which are the price predictors of the house like bedroom count, area in sq feet, distance from school, age of the house etc.

* Vector w contains the unknowns, the model's learnable parameters.

Assuming inverse of ***X*** exists, matrix inversion can solve for w(refer image shown below)

y= ***X*** w

here 'y' values will be  4 and -7 and the matrix can be denoted as two by 1 matrix as shown in the figure.

lets condider the matrix ***X*** as a two by two matrix with values [[4, 2], [-5, -3]].


'w' be the unknown weights, in this case 'b' and 'c'.

From the above pictures it is clear that we can find the unknown values by multiplying **Inverse** of ***X***  with 'y'.

Code implementation for the above problem

```python
X = np.array([[4,2], [-5,-3]])
X
```

```python
# Use linalg.inv() function to find inverse of a matrix
Xinv= np.linalg.inv(X)
Xinv
```

```python
# Solve the unknown weights
y= np.array([4, -7])
y
```

```python
# w = Inverse of X multiplied by vector y
w = np.dot(Xinv, y)
w
```

```python
# Cross-check y=Xw
np.dot(X, w)
```
